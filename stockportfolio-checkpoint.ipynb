```python
import yfinance as yf
# for datetime
import datetime as dt
```

```python
# scrape wiki for individual stock symbols in s&p500
table = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
table = table[0]
```

```python
# rename these 2 companies to yahoo finance symbo format
table['Symbol'] = table['Symbol'].replace({'BRK.B': 'BRK-B'})
table['Symbol'] = table['Symbol'].replace({'BF.B': 'BF-B'})
```

```python
# extract only symbols from table
symbols = table['Symbol'].values.tolist()
```

```python
# get 3 years data
start_date = dt.datetime(2017,1,1)
end_date = dt.datetime(2020,1,4)
```

```python
stockdata=yf.download(symbols,start=start_date,end=end_date)
```

```python
stockdata.head(10)
```

```python
df = pd.DataFrame(stockdata)
```

```python
df = df.reset_index(drop=True)
```

```python
from pypfopt.efficient_frontier import EfficientFrontier
from pypfopt import risk_models
from pypfopt import expected_returns
```

```python
mu=expected_returns.mean_historical_return(df)
S=risk_models.sample_cov(df)

```

```python
#Optimize for the miximal Sharpe ratio
ef = EfficientFrontier(mu, S)
weights = ef.max_sharpe()

cleaned_weights = ef.clean_weights()
print(cleaned_weights)
ef.portfolio_performance(verbose=True)
```
